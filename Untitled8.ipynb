```python
#TASK-4A: Pleasse present a descriptive statistics analysis to provide a general understanding of the dataset. (Describe)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel("Temporary_data3_-_Copy.xlsx")
data
```

```python
#İstatistiksel değerlere ulaşabilmek adına describe() fonksiyonunu kullanabiliriz.
data.describe()
```

```python
#TASK-4B: Please provide the visual illustrations of the dataset that will demonstrate the insightful distribution with respect to the target column (CDRGLOB: the level fo Alzgeimer Disease - AD)
x = data["RSUPTEM"]
y = data["RSUPPAR"]
labels = data["CDRGLOB"]
plt.scatter(x, y, s=6, c=labels, alpha=0.5)
plt.xlabel("RSUPTEM")
plt.ylabel("RSUPPAR")
plt.title("RSUPTEM-RSUPPAR Scatter Plot")
plt.colorbar()
plt.show()
```

```python
x = data ["SMOKYRS"]
y = data["WHITEVOL"]
labels = data["CDRGLOB"]
plt.scatter(x, y, s=10, c=labels, alpha=0.5)
plt.colorbar()
plt.show()
```
